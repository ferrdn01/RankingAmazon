```python
import json
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
with open('amazon_data.json', 'r') as f:
    data = json.load(f)

df = pd.json_normalize(data['data']['deals'], sep='')
df
```

```python
df_dim = df[[
    'product_asin',
    'deal_type',
    'deal_photo',
    'deal_title',
    'deal_state',
    'deal_url',
    'deal_photo'
]]
```

```python
df_fact = df[[
    'product_asin',
    'deal_priceamount',	
    'deal_pricecurrency',
    'list_priceamount',
    'list_pricecurrency',
    'savings_amountamount',
    'savings_amountcurrency',
    'savings_percentage',
    'deal_starts_at',
    'deal_ends_at'
]]
df_fact
```

```python
df_fact = df_fact.rename(columns={
    'deal_priceamount': 'deal_price',
    'list_priceamount': 'list_price',
    'savings_amountamount': 'savings_amount',
    'savings_amountcurrency': 'savings_currency'
}) 

num_cols = ['deal_price', 'list_price', 'savings_amount']
tstp_cols = ['deal_starts_at', 'deal_ends_at']

for col in num_cols:
    df_fact[col] = pd.to_numeric(df_fact[col])

for col in tstp_cols:
    df_fact[col] = pd.to_datetime(df_fact[col])
df_fact
```

```python
num_rows_df = len(df)
num_rows_df_dim = len(df_dim)
num_rows_df_fact = len(df_fact)

print(f"df: {num_rows_df} linhas")
print(f"df_dim: {num_rows_df_dim} linhas")
print(f"df_fact: {num_rows_df_fact} linhas")
```

```python
df_join = pd.merge(df_dim, df_fact, on='product_asin', how='inner')
df_join
```

```python
# Ordena os produtos pelo maior desconto (savings_percentage) e exibe as top 5 sugestões
top_deals = df_join.sort_values(by='savings_percentage', ascending=False).head(5)
display(top_deals[['deal_title', 'deal_price', 'list_price', 'savings_percentage', 'deal_url']])
```

```python


# Extrai o nome reduzido (até a primeira vírgula)
def nome_reduzido(titulo):
    return titulo.split(',')[0] if ',' in titulo else titulo

# Adiciona coluna com nome reduzido
df_join['nome_reduzido'] = df_join['deal_title'].apply(nome_reduzido)

# Identifica os top 5 descontos
top5 = df_join[df_join['top_5']]
outros = df_join[~df_join['top_5']]

# Prepara dados para o gráfico
plt.figure(figsize=(12, 6))
plt.bar(top5['nome_reduzido'], top5['savings_percentage'], color='orange', label='Top 5 descontos')
plt.bar(outros['nome_reduzido'], outros['savings_percentage'], color='gray', alpha=0.5, label='Outros')

plt.ylabel('Desconto (%)')
plt.xlabel('Produto')
plt.title('Comparação dos 5 maiores descontos com os demais')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
